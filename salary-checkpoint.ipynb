```python
df_fl=df2.loc[:,['FL']].sort_values(['FL'], ascending=False)
```

```python
df_ft=df2.loc[:,['FT']].sort_values(['FT'], ascending=False)
```

```python
df_pt=df2.loc[:,['PT']].sort_values(['PT'], ascending=False)
```
